```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('MERGED2017_18_PP.CSV', delimiter = ',')
```

```python
data.head()
```

```python
data.columns
```

```python
null_dict = {}
for col in data.columns:
    null_dict[col] = 0
    for e in data[col]:
        if pd.isnull(e):
            null_dict[col] += 1
```

Now we are going to try and remove most of these columns. 

We will do this by looking at each value in our dictionary of the columns and each of their total number of null values, and only returning the ones that are have below a certain threshold of null values. We'll start with 3500, about 50%.


```python
num_col = 0
cols_half = []
for col in null_dict:
    if null_dict[col] < 3500:
        num_col+=1
        cols_half.append(col)
        
num_col
```

```python
cols_half
```

```python
data[cols_half].to_csv('2017_less_than_half_null.csv')
```
