```python
def getDataFrameFromFiles(path='CollegeScorecard_Raw_Data/'):
    """Creates array of dataframes indexed on year of CollegeScorecard Data

    :param path: Path to unzipped directory of CollegeScorecard data

    :return: an array of data frames indexed by year"""
    
    START_YEAR = 1996
    END_YEAR = 2018
    dfs = {}

    for year in range(START_YEAR, END_YEAR):
        year_YY = year % 100
        dfs[year] = pd.read_csv(path + 'MERGED' + \
                                str(year) + '_{:02d}_PP.CSV'.format((year_YY + 1) % 100), delimiter=',')
    return dfs
```

```python
def addYearAsLabel(dfs):
    for year in dfs:
        df = dfs[year]
        df['Year'] = year
```

```python
def dropUselessColumn(dfs):
    for year in dfs:
        df = dfs[year]
        
        columns_to_drop =   completion_to_drop + \
                            repayment_to_drop + \
                            root_to_drop + \
                            school_to_drop + \
                            earnings_to_drop
        df.drop(set(columns_to_drop).intersection(set(df.columns)), axis=1, inplace=True)
```

```python
def convertUnknownsToNans(dfs):
    df.replace(to_replace='PrivacySuppressed', value=np.nan, inplace=True)
```

```python
def dropColsAllNans(dfs):
    for year in dfs:
        df = dfs[year]
        df.dropna(axis=1, how='all', inplace=True)
        
```

```python
dfs = getDataFrameFromFiles(path='CollegeScorecard_Raw_Data/')
addYearAsLabel(dfs)
```
